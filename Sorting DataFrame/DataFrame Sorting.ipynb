```python
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import rcParams

import seaborn as sb

%matplotlib inline
```

```python
address = 'imdb_1000.csv'
df = pd.read_csv(address, encoding = 'utf-8')
df.head()
```

# Sorting the title series using the sort values method

```python
df['title'].sort_values(ascending = False)
```

```python
type(df['title'])
```

###### Count the numbers of movies by title

```python
df['title'].value_counts()
```

###### We can also use mulple columns to sort the DataFrame

```python
df.sort_values(['title', 'duration', 'genre'], ascending = [0, 1, 1])
```

```python
df.groupby(['genre', 'content_rating']).sum().sort_values('duration', ascending = 1)
```

# Sort the DataFrame by duration using a Condition

We want to sort the DataFrame by duration where genre is Action
Pandas DataFrame.loc attribute access a group of rows and columns by label(s) or a boolean array in the given DataFrame

#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html

```python
df.loc[df['genre'] == 'Action'].sort_values('duration', ascending = 0)
```

###### Sorting DataFrame with groupby method

A groupby operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups.

#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html

```python
df.groupby(['genre']).mean().sort_values('duration', ascending = 0)
```

###### Sorting the DataFrame by duration and Plotting by Genre

```python
rcParams['figure.figsize'] = (13, 10)
sb.set_style('whitegrid')

df.groupby(['genre']).mean().sort_values('duration', ascending = 0).plot(kind = 'barh', color = ['#0394fc', '#c203fc'])

plt.title('Average Star Rating and Duration By Genre', fontdict = {'fontweight':'bold', 'fontsize':14})
plt.ylabel('Movies By Genre', fontdict = {'fontweight':'bold', 'fontsize':14})
plt.xlabel('Ratings and Duration', fontdict = {'fontweight':'bold', 'fontsize':14})
plt.show()
```
