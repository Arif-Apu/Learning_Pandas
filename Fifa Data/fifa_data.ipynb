```python
address = 'fifa_data.csv'
fifa = pd.read_csv(address, encoding = 'utf-8')
fifa.head()
```

###### Filling two or more column's  missing valuees, for an example

```python
fifa['Club'].fillna('No_Club', inplace = True)
```

```python
fifa['Dribbling'].fillna(fifa['Dribbling'].mean(), inplace = True)
```

```python
fifa['Positioning'].fillna(fifa['Positioning'].mean(), inplace = True)
```

```python
fifa['LongPassing'].fillna(fifa['LongPassing'], inplace = True)
fifa['LongShots'].fillna(fifa['LongShots'], inplace = True)
fifa['ShotPower'].fillna(fifa['ShotPower'], inplace = True)
```

```python
fifa['Club'].isnull().sum()
```

###### Visualization

A Histogram for Distribution of Players Overall Skills

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,1])

bins = [40,50,60,70,80,90,100]

ax.set_xticks(bins)
ax.set_title('Distribution of Player Skills')
ax.set_xlabel('Skill Level')
ax.set_ylabel('Number of Players')

color_theme = '#ebc034'
plt.axvline(fifa['Overall'].mean(), linestyle = 'dashed')
ax.hist(fifa.Overall, bins = bins, color = color_theme, alpha = 0.8)
plt.show()
```

Comparison of Potential and Overall Skills

```python
fifa.plot(kind = 'scatter', x = 'Potential', y = 'Overall', c = ['#9e42f5'], s = 20)

plt.title('Potential and Overall Skills of fifa Players', fontsize = 14)
plt.xlabel('Potential', fontsize = 14)
plt.ylabel('Overall', fontsize = 14)
plt.show()
```

Different positions Of the players 

```python
ax = sb.countplot('Position', data = fifa, palette = "Blues_d")
ax.set_xlabel(xlabel = 'Different Positions', fontdict = {'fontsize': 14, 'fontweight':'bold'})
ax.set_ylabel(ylabel = 'Number of Players', fontdict = {'fontsize': 14, 'fontweight':'bold'})
ax.set_title(label = 'An Overview of Players Positions', fontsize = 14)
plt.show()
```

Percentage of Left and Right Foot of the players

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
explode = (0, 0.1)

plt.pie([left, right], labels = labels, autopct = "%.2f%%", colors = ['#4ebfae', '#4ea7bf'],
                shadow=True, startangle = 90, explode=explode)

plt.title('Foot Preferences of FIFA Players')
plt.legend(loc="best", bbox_to_anchor=(0.8, 1))
plt.axis('equal')
plt.show()
```

best players per each position with their age, club, and nationality

```python
fifa.iloc[fifa.groupby(fifa['Position'])['Overall'].idxmax()][['Position', 'Name', 'Age', 'Club', 'Nationality']].head()
```

Total Number of Players By Age

```python
ax = sb.countplot(x = 'Age', data = fifa, palette = 'rainbow')
ax.set_title(label = 'Number of players According to Age', fontsize = 14)
ax.set_xlabel(xlabel = 'Age', fontdict = {'fontsize': 14, 'fontweight':'bold'})
ax.set_ylabel(ylabel = 'Number', fontdict = {'fontsize': 14, 'fontweight':'bold'})
plt.show()
```

A comparison of Shot power and Long shots of all players

```python
shotpower = fifa.groupby('Height').ShotPower.agg(['sum'])
longshots = fifa.groupby('Height').LongShots.agg(['sum'])

plt.scatter(x = shotpower, y = longshots, marker = 's', color = '#bf4ea7')
plt.title('Comparison of Shot power and Long shots By Height', fontsize = 14)
plt.xlabel('Shot Power')
plt.ylabel('Long Shots')
plt.show()
```

Highest number of players of different 8 Countries

```python
colors = ['#bf4ea7','#884ebf', '#4ebf85', '#52bf4e', '#bf614e', '#bfb74e', '#bf834e', '#bf4e57']

fifa['Nationality'].value_counts().head(8).plot(kind = 'bar',  color = colors)

plt.title('Highest Number of Players By Nationality', fontsize = 14)
plt.ylabel('Total Number', fontsize = 14)
plt.xlabel('Country', fontsize = 14)
plt.show()
```

Category of players according to Weight group

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

light = fifa.loc[(fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight > 200)].count()[0]

weights = [light, medium, medium_heavy, heavy]

ax.set_title('Weight Category of FIFA Players (in lbs)')
ax.set_ylabel('')
ax.set_xlabel('')
labels = ['Light(Under 150)', 'Medium(150-175)', 'Medium_heavy(175-200)', 'Heavy(Over 200)']
colors = ['#32a852', '#9ca832', '#c88cd4', '#8c2b34']
explode = (0, 0, 0, 0.4)

ax.pie(weights, labels = labels, autopct = '%.2f%%', pctdistance = 0.5, shadow = False,
      colors = colors,  textprops={'fontsize': 14}, explode = explode)
ax.axis('equal')
ax.legend(loc = 'best', bbox_to_anchor=(1.1, 1))
plt.show()
```

Overall scores of players from Top 8 FIFA ranking Countries

```python
fifa_ranking = ('Belgium', 'France', 'Brazil', 'England', 'Uruguay', 'Croatia', 'Portugal', 'Spain')
fifa_countries = fifa.loc[fifa['Nationality'].isin(fifa_ranking) & fifa['Overall']]

ax = sb.barplot(x = fifa_countries['Nationality'], y = fifa_countries['Overall'], palette = 'ocean')
ax.set_xlabel(xlabel = 'Countries', fontsize = 14)
ax.set_ylabel(ylabel = 'Overall Scores', fontsize = 14)
ax.set_title(label = 'Distribution of Overall scores of players from Top 8 Fifa ranking Countries', fontsize = 20)
plt.show()
```

A Comparison of Top 4 Teams

```python
juventus = fifa.loc[fifa.Club == 'Juventus']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
psg = fifa.loc[fifa.Club == 'Paris Saint-Germain']['Overall']

clubs = [juventus, madrid, barcelona, psg]

labels = ['Juventus', 'Real Madrid', 'FC Barelona', 'Paris Saint-Germain']
plt.title('UEFA Overall Comparison of Top 4 Teams')
plt.ylabel('Overall Rating')

boxes = plt.boxplot(clubs, labels = labels, notch = False, patch_artist = True)

for box in boxes['boxes']:
    #Edge Color
    box.set(color='#4286f4', linewidth = 4)
    #Fill Color
    box.set(facecolor='#a966bd')
    
plt.show()
```
