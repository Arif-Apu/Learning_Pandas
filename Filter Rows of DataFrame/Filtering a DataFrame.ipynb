```python
import pandas as pd
df = pd.read_excel('sales_table.xlsx')
df.head()
```

```python
df.shape
```

Creating a column Named as Profit

```python
df['Profit'] = df.Revenue - df.Cost
```

```python
df.head()
```

We want to see the Data where the Profit is more than 100 

```python
df[df.Profit >= 100]
```

###### Lets look at the Product Category Where Profit is more than 100

```python
df.loc[df['Profit'] > 100, ['Product Category', 'Profit']].sort_values('Product Category')
```

```python
df.loc[df['Quantity'] >= 3, ['Profit', 'Product Category']]
```

```python
df[df.Cost >= 30]['Profit']
```

```python
df.loc[df.Cost >= 40, ['Profit', 'Product Category']]
#df.loc[df['Cost'] >= 40, ['Profit']]
```

###### Multiple Filter Criteria

```python
df.loc[(df['Cost'] >= 50) & (df['Profit'] >= 100), ['Cost', 'Profit', 'Product Category']].sort_values('Product Category')
```

###### Simple Plotting

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
plt.style.use('fivethirtyeight')

%matplotlib inline
```

```python
rcParams['figure.figsize'] = (10, 6)

cost = df.groupby('Product Category').Cost.mean()
revenue = df.groupby('Product Category').Revenue.mean()
profit = df.groupby('Product Category').Profit.mean()

plt.plot(cost, '.-', label = 'Cost')
plt.plot(revenue, '.-', label = 'Revenue')
plt.plot(profit, '.-', label = 'Profit')

plt.title('Total Cost, Revenue and Profit By Year')
plt.xlabel('Product Category')
plt.ylabel('Total in US Dollars')
plt.legend()
plt.show()
```
