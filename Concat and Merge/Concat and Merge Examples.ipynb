```python
import pandas as pd
from pandas import DataFrame, Series

weather1 = pd.DataFrame({
    'city': ['Berlin', 'Paris', 'Rome', 'Prague'],
    'humidity' : [56, 50, 55, 48],
    'temperature' : [17, 18, 23, 19]
    })
weather1
```

```python
weather2 = pd.DataFrame({
    'city' : ['Hamburg', 'Cologne', 'Milan'],
    'humidity': [52, 54, 48],
    'temperature' : [19, 16, 21]
})
weather2
```

###### Concat two dataframes

```python
df = pd.concat([weather1, weather2])
df
```

It remains the original index, To make it as continous index
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html

```python
df = pd.concat([weather1, weather2], ignore_index = True, keys = None, axis = 1)
df
```

```python
df = pd.concat([weather1, weather2], ignore_index = True, keys = None, axis = 0)
df
```

We can associate a key with each of these data frames

```python
df = pd.concat([weather1, weather2], keys = ['Capitals', 'Other Cities'])
df
```

retreive subsets of the dataframes

```python
df.loc['Capitals']
```

###### Merging DataFrame

# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html?highlight=merge#pandas.DataFrame.merge

```python
temperature = pd.DataFrame({
    'city' : ['Berlin', 'Rome', 'Paris', 'Prague'],
    'temperature' : [15, 19, 18, 22]
})
temperature
```

```python
humudity = pd.DataFrame({
    'city' : ['Berlin', 'Rome', 'London'],
    'humudity' : [52, 56, 51]
})
humudity
```

```python
weather = pd.merge(temperature, humudity, on = 'city', how = 'outer')
weather
```

```python
weather = pd.merge(temperature, humudity, on = 'city', how = 'left')
weather
```

```python
weather = pd.merge(temperature, humudity, on = 'city', how = 'inner')
weather
```
