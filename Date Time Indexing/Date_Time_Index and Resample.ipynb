###### Convenience method for frequency conversion and resampling of time series
https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.resample.html

```python
climate['humidity'].resample('M').mean()
```

###### Visualize Average humidity By Monthly

###### https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#time-date-components

```python
climate['humidity'].resample('M').mean().plot( color = ['#fc034e'])

plt.title('Average humidity Over Time (Monthly Frequency))', fontsize = 14)
plt.xlabel('Date')
plt.show()
```

###### Visualizing Data of May Month

```python
climate["2017-05"]['meantemp'].plot()

plt.title('Mean Temp. of May Month')
plt.show()
```
