```python
import pandas as pd

address = 'http://bit.ly/uforeports'
ufo = pd.read_csv(address, encoding = 'utf-8')
ufo.head()
```

###### loc is selecting rows by label position

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html

```python
# 0 label rows and All Columns
#type(ufo.loc[0, :])

ufo.loc[0, :]
```

```python
type(ufo.loc[0, :])
```

```python
ufo.loc[2, :]
```

```python
# row label 2 and State column

ufo.loc[2, 'State']
```

```python
# Multiple Rows and All columns

ufo.loc[[0, 1, 2], :]
```

```python
# Multiple rows from 0-2 and All Columns

ufo.loc[0:2, :]
```

```python
ufo.loc[:, 'City']
```

```python
ufo.loc[:, ['City', 'State']]
```

```python
ufo.loc[:, 'City':'State']
```

```python
ufo.loc[0:9, 'City':'State']
```

```python
ufo.head().drop('Time', axis=1)
```

```python
ufo[ufo.City == 'Oakland']
```

```python
# Rows contains Oakland with all columns

ufo.loc[ufo.City== 'Oakland', :]
```

```python
# Rows contains Oakland and State column

ufo.loc[ufo.City=='Oakland', 'State']
```

```python
ufo[ufo.City=='Oakland'].State
```

#### iloc is filtering rows and selecting columns by integer position

The iloc indexer for Pandas Dataframe is used for integer-location based indexing / selection by position.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iloc.html

```python
ufo.head()
```

```python
# iloc is filtering rows and selecting columns by integer position.
# position 0 and position 3 dataFrame

ufo.iloc[:, [0,3]]
```

```python
#Selecting a range rows/columns. iloc exclusive the last one, but loc inclusive the last one

ufo.iloc[:, 0:4]
```

```python
ufo.iloc[0:3, :]
```

```python
ufo[0:2]
```

###### The ix[] indexer is a hybrid of .loc and .iloc. Generally, ix is label based and acts just as the .loc indexer. 

https://pandas.pydata.org/pandas-docs/version/0.18/generated/pandas.DataFrame.ix.html
    
https://pandas.pydata.org/pandas-docs/version/0.18/advanced.html#advanced    

```python
drinks.ix['Austria', 2]
```

```python
drinks.ix[4, 'wine_servings']
```

```python
drinks.ix['Albania':'Andorra', 0:2]
```

```python
# columns index treate as dot columns and exclusive last one, so get 2 columns.

drinks.ix[0:5, 0:2]
```
