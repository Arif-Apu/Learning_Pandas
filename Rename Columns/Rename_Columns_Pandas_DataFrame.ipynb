###### Here I want to rplace the columns name spaces with underscore. There are several ways to do this. For an example using rename method and pass a dictionary to columns argument. 


```python
df.rename(columns = {'Colors Reported':'Colors_reported', 'Shape Reported':'Shape_Reported'}, inplace =True)
```

```python
df.head()
```

```python
df.columns
```

Another way is to create a python list and overwrite the columns attributes

```python
df_cols = ['city', 'colors_reported', 'shape_reported', 'state', 'time']
df.columns = df_cols
df.head()
```

```python
df.columns
```

But I prefere the replace method to replace columns spaces with underscore

```python
df.columns = df.columns.str.replace(' ', '_')
df
```

We can use lower method to make the columns name Lower Case

```python
df.columns = df.columns.str.lower()
```

```python
df.head()
```
