###### Repalcing Columns name spaces

```python
df.columns = df.columns.str.replace(' ', '')
```

```python
df.head()
```

###### Hndling and Cleaning Missing Data

```python
df.isnull().sum()
```

###### Dropping NaN Values

```python
df = df.dropna(how = 'all')
df.head()
```

```python
df.shape
```

```python
df = df[df['OrderDate'].str[0:2] != 'Or']
```

###### Adding Month Column by astype method

DataFrame.astype() method is used to cast a pandas object to a specified dtype

# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.astype.html

```python
df['Month'] = df['OrderDate'].str[0:2]
df['Month'] = df['Month'].astype('int32')
df.head()
```

###### Dropping a Column

```python
df.drop('Month', axis = 1, inplace = True)
df.head()
```

###### Convert argument to datetime
# https://pandas.pydata.org/pandas-docs/version/0.13/generated/pandas.to_datetime.html

```python
df['OrderDate'] = pd.to_datetime(df.OrderDate)
```

```python
type(df.OrderDate[0])
```

```python
df.OrderDate.dt.weekday_name.head(10)
```

```python
df.to_csv('salesdata.csv', index = False)
```

###### Adding Months Column

```python
df['Months'] = df.OrderDate.dt.month
df.head()
```

###### Convert argument to a numeric type.

The default return dtype is float64 or int64 depending on the data supplied. Use the downcast parameter to obtain other dtypes.

# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_numeric.html

```python
# Make it int
df ['QuantityOrdered'] = pd.to_numeric(df['QuantityOrdered'])
```

```python
type(df.QuantityOrdered[0])
```

```python
# Make it float
df ['PriceEach'] = pd.to_numeric(df['PriceEach'])
```

```python
type(df.PriceEach[0])
```

# Task 1: What was the best month for Total sales? 

```python
df['sales'] = df['QuantityOrdered'] * df['PriceEach']
df.head()
```

###### Reorder Column

```python
df = df.reindex(columns = ['OrderID', 'Product', 'QuantityOrdered', 'PriceEach', 'sales', 
                           'OrderDate', 'PurchaseAddress', 'Months'])
df.head()
```

```python
df.groupby('Months').sales.sum()
```

```python
df.groupby('Months').sales.sum().plot(kind = 'bar')
plt.show()
```

###### Getting the Month Name instead of Month Number

```python
df['MonthName'] = df.OrderDate.dt.strftime('%B') 
df.head()
```

```python
monthname = df.groupby('MonthName').sales.sum().sort_values(ascending = False)

monthname.plot(kind = 'bar', rot = 0, color = ['#ff9999'])
plt.title('Total sales By Months', fontsize = 14, fontweight = 'bold')
plt.xlabel('Month Name', fontsize = 14)
plt.ylabel('Sales in (€)', fontsize = 14)
plt.show()
```

# Task 2: Which Cities are the Top by Total sales?

# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html

# https://pandas.pydata.org/pandas-docs/version/0.18/generated/pandas.Series.apply.html

```python
df['PurchaseAddress'].apply(lambda x: x.split(',')[1])
```

###### Adding city column

```python
df['City'] = df['PurchaseAddress'].apply(lambda x: x.split(',')[1])
df.head()
```

```python
cities = df.groupby('City').sales.sum().sort_values(ascending = True)

cities.plot(kind = 'barh', color = ['#bf00ff'])
plt.title('Total Sales By City', fontsize = 14, fontweight = 'bold')
plt.xlabel('Total Sales in(€)', fontsize = 14)
plt.ylabel('City Name')
plt.show()
```

# Task 3: Which State has the maximum Total sales?

###### Extract The State And Adding a Column

```python
df['State'] = df['PurchaseAddress'].apply(lambda x: x.split(',')[2].split(' ')[1])
df.head()
```

```python
df.groupby('State').sales.sum()
```

```python
grp_state = df.groupby('State').sales.sum().sort_values(ascending = False)

grp_state.plot(kind = 'bar', color = ['#004d4d'], rot = 45)
plt.title('Total Sales By State', fontsize = 14)
plt.xlabel('State Name')
plt.ylabel('Sales in (€)')
plt.show()
```

# Task 4: Find out the time to display advertisements to maximize likelihood of customer's buying product?

```python
df['Hour'] = df['OrderDate'].dt.hour
df.head()
```

```python
df.groupby('Hour').sales.sum()
```

```python
grp_hour = df.groupby('Hour').sales.sum()

grp_hour.plot(color = '#006633')
plt.title('Total Sales Over Hour')
plt.ylabel('Sales in (€)')
plt.show()
```

###### We can also Count the total number of Order Id Over Time To find the total sales

```python
df.groupby('Hour').OrderID.count().plot()
plt.show()
```

# Task 5: What time we had the maximum sales of customer's buying product?

```python
df.groupby('Hour').sales.agg(['max']).plot(color = '#4d4d00')

plt.title('Maximum Sales Over Time(Hour)')
plt.ylabel('Sales in (€)')
plt.show()
```

# Task 6: Find out Which day we had the best total sales ?

```python
df['Day'] = df['OrderDate'].dt.weekday_name
df.head()
```

```python
df.groupby('Day').sales.sum()
```

```python
df.groupby('Day').sales.sum().sort_values(ascending = False).plot(kind = 'bar', rot = 40, color = '#805500')

plt.title('Total Sales By Days')
plt.ylabel('Sales in (€)')
plt.show()
```

# Task 7: What products had the maximum total sales ?

```python
df.groupby('Product').sales.count().sort_values(ascending = False)
```

```python
product = df.groupby('Product').sales.count().sort_values(ascending = False)

product.plot(kind = 'bar', rot = 60, color = '#004d00')
plt.title('Total Sales By Products')
plt.ylabel('Sales in (€)')
plt.show()
```

# Task 8: Find out the Maximum and Minimum Sales for each Product?

```python
df.groupby('Product').sales.min().sort_values(ascending = False)
```

```python
df.groupby('Product').sales.max().sort_values(ascending = False)
```

```python
df.groupby('Product').sales.agg(['max', 'min']).plot(kind = 'bar', stacked = True, rot = 65, color = ['#661a00', '#004d66'])

plt.show()
```

# Task 9: Which Products are Ordered mostly?

```python
df.groupby('Product').QuantityOrdered.sum().sort_values(ascending = False)
```

```python
pro_quantity = df.groupby('Product').QuantityOrdered.sum().sort_values(ascending = False)

pro_quantity.plot(kind = 'bar', rot = 65, color = '#ff8c66')
plt.title('Number of Ordered Quantity By Products')
plt.ylabel('Order Quantity')
plt.xlabel('Product Name')
plt.show()
```
